```python
import jieba
from jieba import lcut, analyse, Tokenizer
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
def Keyword_Finder(keywords):
    for item in keywords:
        print(item[0], item[1])
def Openfile(path):
    f = open(path, "r", encoding="utf-8")
    result = f.read()
    f.close()
    return result
def Openfile_sp(path):
    f = open(path, "r", encoding="utf-8")
    result = f.readlines()
    f.close()
    return result
```

```python
stopwords_file = 'datafile/stop_words.txt'
jieba.set_dictionary('./dict.txt.big.txt')
jieba.load_userdict('datafile/user_dict.txt')
jieba.analyse.set_stop_words('datafile/stop_words.txt')
```

```python
#cut to cloud
dec = Openfile_sp(stopwords_file)
stopwords = set(STOPWORDS)
for i in range(len(dec)):
    stopwords.add(dec[i].strip())
```

```python
target_file = 'datafile/word_cloud/EVA/bef2017.txt'
target_text = Openfile(target_file)
```

```python
cut1 = jieba.cut(target_text, cut_all=True, HMM=False)
cut1_sp = " ".join(cut1)
```

```python
target_cloud = WordCloud(
    font_path="msj.ttf", #設置字體
    background_color="white", #背景顏色
    stopwords=stopwords,
    max_words=100)
```

```python
target_cloud.generate(cut1_sp)
```

```python
plt.imshow(target_cloud)
plt.axis("off")
plt.figure(figsize=(10,6), dpi=400)
plt.savefig('image/test.png')
plt.show()
```
